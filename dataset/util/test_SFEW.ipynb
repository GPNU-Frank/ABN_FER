```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import sys
sys.path.append('..')
from SFEW import SFEW
import torch.utils.data as Data
```

```python
root = '../../data/SFEW/Train/Train_Aligned_Faces'

train_set = SFEW(root=root)
```

```python
X, y = [], []
for i in range(10):
    X.append(train_set[i][0])
    y.append(train_set[i][1])

text_label = ['Angry', 'Disgust', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise']

def get_label_name(labels):
    return [text_label[int(i)] for i in labels]
def show_image_grid(images, labels):
    _, figs = plt.subplots(1, len(images), figsize=(12, 12))
    for f, img, lbl in zip(figs, images, labels):
        f.imshow(img)
        f.set_title(lbl)
        f.axes.get_xaxis().set_visible(False)
        f.axes.get_yaxis().set_visible(False)
    plt.show()
```

```python
show_image_grid(X, get_label_name(y))
```
